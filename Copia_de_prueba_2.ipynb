# Ejercicio 1 :Dataframe a partir de la lista 'exam_data'





```python
import pandas as pd

exam_data = {
             'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
             'score': [ 12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19],
             'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
             'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
             }
df = pd.DataFrame(exam_data)
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df.index = labels
df
```

# **1.1**  Seleccione solo las columnas **name** y **score**::

```python
df[['name', 'score']]

```

#**1.2** Filtre las filas donde el número de intentos **attemps** sea **mayor que 2**.


```python
df[df['attempts']>2]
```


#**1.3** Filtre las filas donde el score se encuentre entre **15 y 20 incluidos**

Se utilizó lafunción .between() que se usa para para filtrar filas de un DataFrame en función de si los valores de una columna están dentro de un rango especificado.


```python
df[df.score.between(15,20)]
```

#**1.4**. Cambie el score en la fila 'd' a **11.5**:

Uso de .loc[] para acceder y filtrar datos en un DataFrame de Pandas utilizando etiquetas de índice.


```python
df.loc['d','score'] =11.5
df
```

# **1.5 Agregue** una nueva fila con etiqueta **'k'** al dataframe con un valor que usted desee para cada columna:

```python
df.loc['k'] = {
             'name': 'Andres',
             'score': 17,
             'attempts':2,
             'qualify': 'yes'
}
df
```

#**1.6 Borre** la fila **'k'** y entregue nuevamente el DataFrame original.

Uso de .drop()  para eliminar filas o columnas de un DataFrame.

```python
df= df.drop('k')
df
```

# **1.7 Ordene** el DataFrame por nombre en orden ascendente:

método sort_values() para ordenar un DataFrame o una Serie según los valores de una o varias columnas.


```python
df = df.sort_values(by = 'name', ascending= True)
df
```

# **1.8 Reemplace** los valores de la columna qualify que contiene los valores yes y no por los valores **booleanos** True y False respectivamente:

método .map()  para aplicar una función o un diccionario a cada elemento de una Serie o columna de un DataFrame.

```python
df['qualify']=df['qualify'].map({'yes':True, 'no':False})
df
```

#**3.2 Separe la variable Age en tres grupos y cree una nueva columna llamada Age-Group con esta información**.

se dividió usando la función .cut() y definiendo los rangos de edad y las etiquetas correspondientes ('labels')

```python
df_3['Age_group'] = pd.cut(df_3['Age'], bins = (0, 25, 45, 80), labels= ('Young', 'Adult','Senior'))
df_3.head(30)
```

# 3.3 Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para para los cinco equipos que pagan mejores salarios.

Primero sacamos los equipos top de la lista anterios  y creamos 'equipos_top', luego usamos la función '.isin para encontrar los equipoios que estaban en la variable  'equipos_top' y kuego condtruimos la pivot_table. Siendo 'Chicago Bulls' el equipo con mayor promedio de salarios.

```python
equipos_top=df_3.groupby(['Team'])['Salary'].sum().sort_values(ascending=False).head(5).index

print('lo equipos con más dinero en nomina son:',equipos_top)
```

```python
top=df_3[df_3['Team'].isin(equipos_top)]
```

```python
top.pivot_table(index=['Team','Position'],values='Salary',aggfunc='mean')
```
